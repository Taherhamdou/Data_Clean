```python
df = df.drop(columns=['Junk Column'])
df.head()
#Delete Unwanted column
```

```python
df['Study Hours per Week'] = df['Study Hours per Week'].str.replace(r'[^0-9.-]', '', regex=True)
#remove the strings from the numeric columns
df['Study Hours per Week'] = pd.to_numeric(df['Study Hours per Week'], errors='coerce')

df['Study Hours per Week'] = df['Study Hours per Week'].astype(float)
df['Study Hours per Week'].fillna(df['Study Hours per Week'].mean(), inplace=True)
#convert numeric columns to float
```

```python

data_cleaned = df.drop(columns=['Junk Column'])

# Drop rows with too many missing values
threshold = len(data_cleaned.columns) * 0.6  # Keep rows with at least 60% non-NA values
data_cleaned = data_cleaned.dropna(thresh=threshold)

# Handle categorical inconsistencies (e.g., "Yet" to "Yes")
data_cleaned['Participation in Extracurricular Activities'] = data_cleaned[
    'Participation in Extracurricular Activities'
].replace({'Yet': 'Yes'})

# Impute missing values for numerical columns using median
num_cols = ['Study Hours per Week', 'Attendance Rate', 'Previous Grades']
data_cleaned[num_cols] = data_cleaned[num_cols].fillna(data_cleaned[num_cols].median())

# Drop rows with missing values in categorical columns for simplicity
data_cleaned = data_cleaned.dropna(subset=['Participation in Extracurricular Activities',
                                           'Parent Education Level', 'Passed'])

# Display cleaned data info
cleaned_data_info = {
    "shape": data_cleaned.shape,
    "missing_values": data_cleaned.isnull().sum(),
    "sample_data": data_cleaned.head()
}

cleaned_data_info
```
